## Task 1

```python
# given the data in the task (no output)
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

### 1a) 

```python
# we first filter valid transactions (no output)
valid_transactions = []

for t in transactions:
    if t["price"] > 0 and t["quantity"] >= 0:
        valid_transactions.append(t)
```

### 1b)

```python
# then we compute total = price * quantity (no output)
for t in valid_transactions:
    t["total"] = t["price"] * t["quantity"]
```

### 1c)

```python
# here we group by region and sum totals (no output)
revenue_by_region = {}

for t in valid_transactions:
    region = t["region"]
    
    if region not in revenue_by_region:
        revenue_by_region[region] = 0
    
    revenue_by_region[region] += t["total"]
```

### 1d)

```python
# then find region with highest revenue and finally get some output
max_region = None
max_value = 0

for region, value in revenue_by_region.items():
    if value > max_value:
        max_value = value
        max_region = region

print(f"Region with highest revenue: {max_region} ({max_value})")
```

## Task 2

```python
# given sales data (2 weeks)
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]

# function to summarize the sales data
def summarise_sales(sales):

    # first we compute the total sales
    total = 0
    for value in sales:
        total += value

    # then we calculate the average (rounded to 2 decimals)
    average = round(total / len(sales), 2)

    # find the best and worst day (index starts from 1)
    max_value = sales[0]
    min_value = sales[0]
    best_day = 1
    worst_day = 1

    for i in range(len(sales)):
        if sales[i] > max_value:
            max_value = sales[i]
            best_day = i + 1

        if sales[i] < min_value:
            min_value = sales[i]
            worst_day = i + 1

    # count how many days are above average
    above_average_count = 0
    for value in sales:
        if value > average:
            above_average_count += 1

    # store everything in a dictionary
    result = {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

    return result


# call the function and print the result
summary = summarise_sales(sales)
print(summary)
```

```python
# given student data
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

```python
# creating list of students who passed (grade >= 50)
passing = []

for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print(passing)
```

```python
# b) printing status message for each student
for student in students:
    name = student["name"]
    grade = student["grade"]

    if grade >= 80:
        print(f"{name}: Excellent")
    elif grade >= 50:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")
```

```python
# c) counting students with attendance below 0.80
low_attendance_count = 0

for student in students:
    if student["attendance"] < 0.80:
        low_attendance_count += 1

print(low_attendance_count)
```

### 3d)

```python
# d) finding student with highest grade using a loop
top_student = students[0]

for student in students:
    if student["grade"] > top_student["grade"]:
        top_student = student

print(top_student["name"], top_student["grade"])
```

### 3e)

```python
# e) finding student with highest grade using max()
best = max(students, key=lambda x: x["grade"])

print(best["name"], best["grade"])
```

## Task 4

```python
# Simple class to keep track of a portfolio
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}   

    # add a stock or update it if it already exists
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }

    # go through everything and sum up total value
    def total_value(self):
        total = 0

        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            total += shares * price

        return total

    # find which stock is worth the most in the portfolio
    def most_valuable(self):
        top_ticker = None
        top_value = 0

        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            value = shares * price

            if value > top_value:
                top_value = value
                top_ticker = ticker

        return top_ticker

    # just a clean summary when printing the object
    def __str__(self):
        number_of_assets = len(self.holdings)
        total = self.total_value()

        return f"Portfolio owner: {self.owner}, assets: {number_of_assets}, total value: {total}"
```

### 4a)

```python
# making a portfolio for myself
portfolio = Portfolio("Frans")

# adding some stocks 
portfolio.add_asset("NVDA", 6, 900)
portfolio.add_asset("ASML", 3, 950)
portfolio.add_asset("LVMH", 2, 800)
portfolio.add_asset("AMD", 10, 180)

# updating NVDA to show it overwrites properly
portfolio.add_asset("NVDA", 7, 920)
```

### 4b)

```python
# Get total portfolio value 
print(portfolio.total_value())
```

### 4c)

```python
# Find the most valuable asset
print(portfolio.most_valuable())
```

### 4d)

```python
# Print a summary
print(portfolio)
```

```python
# import the datetime module
from datetime import datetime

# given data
orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),
]

# first we convert all date strings into date objects
dates = []

for order in orders:
    date_string = order[0]
    date_object = datetime.strptime(date_string, "%Y-%m-%d").date()
    dates.append(date_object)

# find earliest and latest dates
earliest = dates[0]
latest = dates[0]

for d in dates:
    if d < earliest:
        earliest = d
    if d > latest:
        latest = d

# compute the difference in days
difference = latest - earliest

# print result
print(difference.days)
```

## Task 8

```python
import numpy as np
import pandas as pd

# generate the data
rng = np.random.default_rng(seed=123)

days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)
```

### 8a)

```python
# create DataFrame and set date as index
df = pd.DataFrame({
    "date": days,
    "visitors": visitors,
    "conversion_rate": conversion_rate,
    "avg_order_value": avg_order_value
})

df = df.set_index("date")

print("Initial DataFrame:\n", df.head(), "\n")
```

### 8b)

```python
# create new columns
# conversions = visitors * conversion_rate (rounded)
df["conversions"] = (df["visitors"] * df["conversion_rate"]).round(0)

# daily revenue = conversions * avg_order_value
df["daily_revenue"] = df["conversions"] * df["avg_order_value"]

print("With new columns:\n", df.head())
```
